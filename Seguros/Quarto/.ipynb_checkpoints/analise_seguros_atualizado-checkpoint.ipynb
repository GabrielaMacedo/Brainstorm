## Introdução
Nesta apresentação analisamos a base de adesão mensal, focando em produtos e perfil do cliente.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuração visual (matplotlib puro)
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.grid"] = True

# ------------------------------------------------------------
# Load dos dados
# - tenta primeiro no diretório do notebook
# - se não encontrar, usa o caminho padrão do ambiente (/mnt/data)
# ------------------------------------------------------------
FILE_LOCAL = "ADESAO_MES.xlsx"
FILE_FALLBACK = "/mnt/data/ADESAO_MES.xlsx"

file_path = FILE_LOCAL if os.path.exists(FILE_LOCAL) else FILE_FALLBACK

df = pd.read_excel(file_path)

# Conversão de datas (se existir no arquivo)
if "DAT_ADESAO" in df.columns:
    df["DAT_ADESAO"] = pd.to_datetime(df["DAT_ADESAO"], errors="coerce")
    df["MES_ANO"] = df["DAT_ADESAO"].dt.to_period("M").astype(str)

print("Dados carregados com sucesso!")
print("Arquivo usado:", file_path)
print("Shape:", df.shape)
df.head(3)
```

```python
# Cálculo de métricas principais
total_adesoes = df.shape[0]
seguradoras_unicas = df['DSC_SEGURADORA'].nunique()
media_idade = df['NUM_IDADE'].mean()
produto_top = df['DSC_PRODUTO'].mode()[0]

print(f"--- Resumo Executivo ---")
print(f"Total de Adesões: {total_adesoes}")
print(f"Seguradoras Parceiras: {seguradoras_unicas}")
print(f"Idade Média do Cliente: {media_idade:.1f} anos")
print(f"Produto Mais Vendido: {produto_top}")
```

## Perfil Demográfico (demografia pura)

Nesta seção analisamos **apenas variáveis demográficas**, sem foco em produto/receita:

- Idade (e faixas)
- Gênero
- Estado civil
- Dependentes
- Tipo de residência
- Profissão (top N)
- Domínio de e-mail (top N)
- Geografia: UF, região e CEP (prefixos)


```python
#| echo: false
# ------------------------------------------------------------
# 1) Seleção + preparação (demografia pura)
# ------------------------------------------------------------
COL_IDADE          = "NUM_IDADE"
COL_GENERO         = "DSC_GENERO_AJUSTADO"
COL_ESTADO_CIVIL   = "DSC_ESTADO_CIVIL"
COL_DEPENDENTES    = "QTD_DEPENDENTES"
COL_TIPO_RESID     = "DSC_TIPO_RESIDENCIA"
COL_PROFISSAO      = "DSC_PROFISSAO"
COL_DOMINIO_EMAIL  = "DSC_DOMINIO_EMAIL"
COL_CEP            = "NUM_CEP"
COL_UF             = "SIG_UF"
COL_REGIAO_SIG     = "SIG_REGIAO"
COL_REGIAO_DESC    = "DSC_REGIAO"

COLS_DEMO = [
    COL_IDADE, COL_GENERO, COL_ESTADO_CIVIL, COL_DEPENDENTES,
    COL_TIPO_RESID, COL_PROFISSAO, COL_DOMINIO_EMAIL,
    COL_CEP, COL_UF, COL_REGIAO_SIG, COL_REGIAO_DESC
]

presentes = [c for c in COLS_DEMO if c in df.columns]
faltantes = [c for c in COLS_DEMO if c not in df.columns]

print("Colunas demográficas presentes:", presentes)
if faltantes:
    print("Colunas demográficas faltantes:", faltantes)

base = df[presentes].copy()

def norm_str(s: pd.Series) -> pd.Series:
    return s.astype("string").str.strip().str.upper()

def normalizar_cep(s: pd.Series) -> pd.Series:
    return (
        s.astype("string")
         .str.replace(r"\.0$", "", regex=True)
         .str.replace(r"\D", "", regex=True)
    )

def normalizar_dominio_email(s: pd.Series) -> pd.Series:
    s = norm_str(s)
    return (
        s.str.replace("HOTMIAL.COM", "HOTMAIL.COM", regex=False)
         .str.replace("GMAIL.COM.BR", "GMAIL.COM", regex=False)
    )

# Tipagens e padronizações
if COL_IDADE in base.columns:
    base[COL_IDADE] = pd.to_numeric(base[COL_IDADE], errors="coerce")

if COL_DEPENDENTES in base.columns:
    base[COL_DEPENDENTES] = pd.to_numeric(base[COL_DEPENDENTES], errors="coerce")

for c in [COL_GENERO, COL_ESTADO_CIVIL, COL_TIPO_RESID, COL_PROFISSAO, COL_UF, COL_REGIAO_SIG, COL_REGIAO_DESC]:
    if c in base.columns:
        base[c] = norm_str(base[c])

if COL_DOMINIO_EMAIL in base.columns:
    base["DOMINIO_EMAIL_NORM"] = normalizar_dominio_email(base[COL_DOMINIO_EMAIL])
else:
    base["DOMINIO_EMAIL_NORM"] = pd.NA

if COL_CEP in base.columns:
    base[COL_CEP] = normalizar_cep(base[COL_CEP])
    base["CEP_2DIG"] = base[COL_CEP].str.slice(0, 2)
    base["CEP_3DIG"] = base[COL_CEP].str.slice(0, 3)
else:
    base["CEP_2DIG"] = pd.NA
    base["CEP_3DIG"] = pd.NA

# Faixas (idade e dependentes)
if COL_IDADE in base.columns:
    base["FX_IDADE"] = pd.cut(
        base[COL_IDADE],
        bins=[17, 24, 34, 44, 54, 64, 120],
        labels=["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
    )

if COL_DEPENDENTES in base.columns:
    base["FX_DEPENDENTES"] = pd.cut(
        base[COL_DEPENDENTES],
        bins=[-0.1, 0, 1, 2, 10],
        labels=["0", "1", "2", "3+"]
    )

print("Base demográfica (shape):", base.shape)
base.head(5)
```

## Qualidade de dados (missing)

## Funções de visualização (matplotlib puro)

```python
#| echo: false
def plot_bar_count(df_, col, top_n=20, title=None):
    vc = df_[col].value_counts(dropna=False).head(top_n)
    plt.figure(figsize=(10, 6))
    vc.sort_values(ascending=True).plot(kind="barh")
    plt.title(title or f"Distribuição: {col}")
    plt.xlabel("Quantidade")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

def plot_hist(df_, col, bins=25, title=None):
    s = df_[col].dropna()
    plt.figure(figsize=(10, 6))
    plt.hist(s, bins=bins)
    plt.title(title or f"Histograma: {col}")
    plt.xlabel(col)
    plt.ylabel("Frequência")
    plt.tight_layout()
    plt.show()

def plot_age_pyramid(df_, age_col, sex_col, bins=None, title="Pirâmide Etária"):
    if bins is None:
        bins = [18, 25, 35, 45, 55, 65, 80]

    sub = df_[[age_col, sex_col]].dropna().copy()
    sub["FAIXA"] = pd.cut(sub[age_col], bins=bins, right=False)
    tab = pd.crosstab(sub["FAIXA"], sub[sex_col]).sort_index()

    # tenta padronizar nomes esperados
    cols = tab.columns.astype(str).tolist()
    fem = "FEMININO" if "FEMININO" in cols else (cols[0] if cols else None)
    mas = "MASCULINO" if "MASCULINO" in cols else (cols[1] if len(cols) > 1 else (cols[0] if cols else None))

    if fem is None or mas is None:
        print("Não foi possível construir pirâmide etária (colunas insuficientes).")
        return

    # garante colunas (para não quebrar)
    if fem not in tab.columns: tab[fem] = 0
    if mas not in tab.columns: tab[mas] = 0

    y = np.arange(len(tab.index))
    f = tab[fem].values
    m = tab[mas].values

    plt.figure(figsize=(10, 6))
    plt.barh(y, -m, label=str(mas))
    plt.barh(y,  f, label=str(fem))
    plt.yticks(y, tab.index.astype(str))
    plt.title(title)
    plt.xlabel("Quantidade (masc. à esquerda, fem. à direita)")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## Idade

```python
#| echo: false
if COL_IDADE in base.columns:
    display(base[COL_IDADE].describe(percentiles=[0.01,0.05,0.25,0.5,0.75,0.95,0.99]).to_frame().T)
    plot_hist(base, COL_IDADE, bins=25, title="Distribuição de Idade")
    if "FX_IDADE" in base.columns:
        plot_bar_count(base, "FX_IDADE", top_n=20, title="Faixa Etária")
else:
    print("Coluna de idade não encontrada.")
```

## Gênero

```python
#| echo: false
if COL_GENERO in base.columns:
    plot_bar_count(base, COL_GENERO, top_n=20, title="Distribuição por Gênero")
    if COL_IDADE in base.columns:
        plot_age_pyramid(base, COL_IDADE, COL_GENERO, title="Pirâmide Etária (por gênero)")
else:
    print("Coluna de gênero não encontrada.")
```

## Estado civil e dependentes

```python
#| echo: false
if COL_ESTADO_CIVIL in base.columns:
    plot_bar_count(base, COL_ESTADO_CIVIL, top_n=15, title="Estado civil (top 15)")
else:
    print("Coluna de estado civil não encontrada.")

if "FX_DEPENDENTES" in base.columns:
    plot_bar_count(base, "FX_DEPENDENTES", top_n=10, title="Dependentes (faixas)")
elif COL_DEPENDENTES in base.columns:
    plot_hist(base, COL_DEPENDENTES, bins=15, title="Distribuição de dependentes")
else:
    print("Coluna de dependentes não encontrada.")
```

## Residência, profissão e domínio de e-mail

```python
#| echo: false
if COL_TIPO_RESID in base.columns:
    plot_bar_count(base, COL_TIPO_RESID, top_n=12, title="Tipo de residência (top 12)")
else:
    print("Coluna de tipo de residência não encontrada.")

if COL_PROFISSAO in base.columns:
    plot_bar_count(base, COL_PROFISSAO, top_n=20, title="Profissão (top 20)")
else:
    print("Coluna de profissão não encontrada.")

if "DOMINIO_EMAIL_NORM" in base.columns:
    plot_bar_count(base, "DOMINIO_EMAIL_NORM", top_n=15, title="Domínio de e-mail (top 15)")
```

## Geografia: UF, região e CEP

```python
#| echo: false
if COL_UF in base.columns:
    plot_bar_count(base, COL_UF, top_n=27, title="Distribuição por UF")
else:
    print("Coluna de UF não encontrada.")

if COL_REGIAO_DESC in base.columns:
    plot_bar_count(base, COL_REGIAO_DESC, top_n=10, title="Distribuição por Região")
elif COL_REGIAO_SIG in base.columns:
    plot_bar_count(base, COL_REGIAO_SIG, top_n=10, title="Distribuição por Região (sigla)")
else:
    print("Coluna de região não encontrada.")

if "CEP_2DIG" in base.columns:
    plot_bar_count(base, "CEP_2DIG", top_n=25, title="CEP (prefixo 2 dígitos) — top 25")
```

## Conclusões

- **Completude**: priorize tratamento de missing nas variáveis com maior ausência antes de segmentações.
- **Núcleo demográfico**: idade/faixa etária + gênero + estado civil + dependentes capturam o perfil base.
- **Alta cardinalidade**: profissão e CEP devem ser comunicados por **top N** e/ou **prefixos**.
