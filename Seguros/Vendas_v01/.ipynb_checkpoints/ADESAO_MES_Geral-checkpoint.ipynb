## 2. Analise das Adesoes por Variável Categorica

```python
# Converter a coluna ANO_MES_ADESAO_SEGURO para string (caso esteja como int)
df['ANO_MES_ADESAO_SEGURO'] = df['ANO_MES_ADESAO_SEGURO'].astype(str)

# Agrupar e contar os ID_ADESAO por mês
df_grouped = df.groupby('ANO_MES_ADESAO_SEGURO')['ID_ADESAO'].count().reset_index()

# Ordenar cronologicamente
df_grouped = df_grouped.sort_values('ANO_MES_ADESAO_SEGURO')
```

```python
def plot_personalizado(
    cor_linha='#522FA7',
    cor_marcador='#522FA7',
    cor_texto='#414141',
    cor_grid='white',
    cor_borda_plot='white',
    cor_fundo_plot='white',
    salvar_como='grafico_adesoes.png'  # caminho do arquivo final
):
    x = np.arange(len(df_grouped))
    y = df_grouped['ID_ADESAO'].values
    labels = df_grouped['ANO_MES_ADESAO_SEGURO']
    x_smooth = np.linspace(x.min(), x.max(), 300)
    spline = make_interp_spline(x, y, k=3)
    y_smooth = spline(x_smooth)

    limite_superior = y.max() * 1.3

    fig, ax = plt.subplots(figsize=(12, 6))

    # Plotagem da linha e marcadores
    ax.plot(x_smooth, y_smooth, color=cor_linha, linewidth=2)
    ax.scatter(x, y, color=cor_marcador, edgecolors=cor_marcador, s=60, zorder=3)

    # Eixos
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, color=cor_texto)
    ax.set_xlabel('Ano/Mês da Adesão', color=cor_texto)
    ax.set_ylabel('Quantidade de Adesões', color=cor_texto)
    ax.set_ylim(bottom=0, top=limite_superior)

    # Título
    ax.set_title('Adesões por Mês', fontsize=14, loc='left', color=cor_texto)

    # Grade
    ax.grid(True,color=cor_grid)

    # Cor de fundo da área de plotagem
    ax.set_facecolor(cor_fundo_plot)

    # Cor da borda da área de plotagem
    for spine in ax.spines.values():
        spine.set_color(cor_borda_plot)

    # Salvar imagem
    plt.tight_layout()
    plt.savefig(salvar_como, dpi=300, bbox_inches='tight')
    plt.show()

# 5. Exemplo de uso
plot_personalizado(
    cor_linha='#522FA7',
    cor_marcador='#522FA7',
    cor_texto='#414141',
    cor_grid='#f3f3f3',
    cor_borda_plot='#414141',
    cor_fundo_plot='white',
    salvar_como='grafico_adesoes_customizado.png'
)
```

```python
def plot_subplots_variavel_categorica(
    df,
    variavel_categoria,
    lista_cores,
    colunas=3,
    cor_texto='#414141',
    cor_grid='#f3f3f3',
    cor_borda_plot='#414141',
    cor_fundo_plot='white',
    salvar_como=None  # agora é opcional
):
    # 0. Gera nome automático se não for fornecido
    if salvar_como is None:
        salvar_como = f'adesoes_por_{variavel_categoria.lower()}.png'

    # 1. Identificar categorias únicas
    categorias = sorted(df[variavel_categoria].dropna().unique())
    qtd_categorias = len(categorias)

    # 2. Validar lista de cores
    if qtd_categorias > len(lista_cores):
        raise ValueError(f"Número de categorias ({qtd_categorias}) excede o número de cores disponíveis ({len(lista_cores)}).")

    # 3. Mapear categorias para cores
    cores_utilizadas = {cat: lista_cores[i] for i, cat in enumerate(categorias)}

    # 4. Agrupamento
    df_grouped = df.groupby([variavel_categoria, 'ANO_MES_PORTFOLIO'])['ID_ADESAO'].count().reset_index()
    max_adesoes = df_grouped['ID_ADESAO'].max() * 1.2
    linhas = math.ceil(qtd_categorias / colunas)

    # 5. Criar subplots
    fig, axes = plt.subplots(linhas, colunas, figsize=(6 * colunas, 4 * linhas), sharey=True)
    axes = axes.flatten()

    for i, (categoria, cor) in enumerate(cores_utilizadas.items()):
        df_categoria = df_grouped[df_grouped[variavel_categoria] == categoria]
        x = np.arange(len(df_categoria))
        y = df_categoria['ID_ADESAO'].values
        labels = df_categoria['ANO_MES_PORTFOLIO']

        # Linha de tendência
        if len(x) > 1:
            coef = np.polyfit(x, y, 1)
            y_trend = np.poly1d(coef)(x)
        else:
            y_trend = y  # apenas repete valor

        # Linha principal
        axes[i].plot(x, y, color=cor, linewidth=2)
        axes[i].scatter(x, y, color=cor, edgecolors=cor, s=40, zorder=3)

        # Linha de tendência pontilhada
        axes[i].plot(x, y_trend, color=cor, linestyle='--', linewidth=2, alpha=0.8)

        # Ajustes
        axes[i].set_xticks(x)
        axes[i].set_xticklabels(labels, rotation=45, fontsize=9, color=cor_texto)
        axes[i].set_title(categoria, loc='left', fontsize=12, color=cor_texto)
        axes[i].set_ylim(0, max_adesoes)
        axes[i].set_facecolor(cor_fundo_plot)
        for spine in axes[i].spines.values():
            spine.set_color(cor_borda_plot)
        axes[i].grid(True, color=cor_grid, alpha=0.9)

    # 6. Desativar eixos não usados
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    fig.suptitle(f'Adesões por Categoria: {variavel_categoria}', fontsize=16, x=0.01, ha='left', color=cor_texto)
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.savefig(salvar_como, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def plot_barras_empilhadas_100pct_multivariaveis(
    df,
    variaveis_categoricas,
    lista_cores,
    colunas=2,
    cor_texto='#414141',
    salvar_como='barras_empilhadas_100pct.png'
):
    linhas = math.ceil(len(variaveis_categoricas) / colunas)
    fig, axes = plt.subplots(linhas, colunas, figsize=(7 * colunas, 5 * linhas))
    axes = axes.flatten()

    for idx, var in enumerate(variaveis_categoricas):
        # Agrupar
        df_grouped = df.groupby([df['ANO_MES_PORTFOLIO'], df[var]])['ID_ADESAO'].count().reset_index()
        df_pivot = df_grouped.pivot(index='ANO_MES_PORTFOLIO', columns=var, values='ID_ADESAO').fillna(0)

        # Calcular proporção 100%
        df_pct = df_pivot.div(df_pivot.sum(axis=1), axis=0) * 100

        # Garantir ordem das cores
        categorias = df_pct.columns.tolist()
        if len(categorias) > len(lista_cores):
            raise ValueError(f"Número de categorias em {var} excede o número de cores disponíveis.")

        cores_dict = {cat: lista_cores[i] for i, cat in enumerate(categorias)}

        # Plotagem
        df_pct.plot(
            kind='bar',
            stacked=True,
            ax=axes[idx],
            color=[cores_dict[cat] for cat in categorias],
            width=0.9
        )

        axes[idx].set_title(f'{var}', loc='left', fontsize=12, color=cor_texto)
        axes[idx].set_ylabel('Percentual (%)')
        axes[idx].set_xlabel('')
        axes[idx].set_ylim(0, 100)
        axes[idx].legend(title=var, bbox_to_anchor=(1.05, 1), loc='upper left')
        axes[idx].tick_params(axis='x', rotation=45)
        axes[idx].grid(axis='y', linestyle='--', alpha=0.4)

    # Desativar eixos não usados
    for i in range(len(variaveis_categoricas), len(axes)):
        axes[i].axis('off')

    plt.tight_layout()
    plt.savefig(salvar_como, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
variaveis_categoricas = [
      'DSC_TIPO_PRODUTO'
    , 'DSC_EMPRESA'
    , 'DSC_TIPO_VENDA'
    , 'DSC_SEGURADORA'
    , 'DSC_GRP_CANAL_VENDA_SEG'
    , 'QTD_SEGS_VIGENTES'
    , 'FLG_EXCL_PRESTAMISTA'
    , 'DSC_RATING_AQUISICAO_STR'
    , 'DSC_RATING_BACEN_STR'
    , 'DSC_RATING_STR'
    , 'FX_LIMITE'
    , 'DSC_ESTADO_CIVIL'
    , 'FAIXA_IDADE_LABEL'
    , 'DSC_DOMINIO_EMAIL'
    , 'DSC_GENERO_AJUSTADO'
    , 'DSC_REGIAO'
]

plot_barras_empilhadas_100pct_multivariaveis(
    df=df,
    variaveis_categoricas=variaveis_categoricas,
    lista_cores=lista_cores,
    salvar_como='graficos_empilhados_variaveis.png'
)
```

```python
def identificar_persona_mensal(df, lista_variaveis, coluna_tempo='ANO_MES_PORTFOLIO', coluna_id='ID_ADESAO'):
    resultados = []

    # Loop por mês
    for mes, df_mes in df.groupby(coluna_tempo):
        for var in lista_variaveis:
            if var not in df_mes.columns:
                continue  # ignora variáveis ausentes

            contagem = df_mes[var].value_counts(dropna=False)
            total = contagem.sum()

            categoria_top = contagem.idxmax()
            frequencia_top = contagem.max()
            perc_top = round((frequencia_top / total) * 100, 1)

            resultados.append({
                'ANO_MES_PORTFOLIO': mes,
                'Variável': var,
                'Categoria mais frequente': categoria_top,
                'Frequência (%)': f'{perc_top}%'
            })

    return pd.DataFrame(resultados)
```

```python
variaveis_persona = [
    'DSC_TIPO_PRODUTO', 'DSC_EMPRESA', 'DSC_TIPO_VENDA', 'DSC_SEGURADORA',
    'DSC_GRP_CANAL_VENDA_SEG', 'QTD_SEGS_VIGENTES', 'FLG_EXCL_PRESTAMISTA',
    'DSC_RATING_AQUISICAO_STR', 'DSC_RATING_BACEN_STR', 'DSC_RATING_STR',
    'FX_LIMITE', 'DSC_ESTADO_CIVIL', 'FAIXA_IDADE_LABEL', 'DSC_DOMINIO_EMAIL',
    'DSC_GENERO_AJUSTADO', 'DSC_REGIAO'
]

df_personas_mensal = identificar_persona_mensal(df, variaveis_persona)
display(df_personas_mensal)
```

```python
def criar_tabela_comparativa_categorias(df_resultado_personas):
    # Pivotar a tabela: variáveis como linhas, meses como colunas, valores = categoria mais frequente
    tabela_comparativa = df_resultado_personas.pivot(
        index='Variável',
        columns='ANO_MES_PORTFOLIO',
        values='Categoria mais frequente'
    )

    # Ordenar colunas por tempo (opcional, se ANO_MES for string YYYY-MM)
    tabela_comparativa = tabela_comparativa[sorted(tabela_comparativa.columns)]

    return tabela_comparativa
```

## 🔶 1️⃣ Análise de tendência (trend analysis)
Use modelos simples como regressão linear para ver se há tendência crescente ou decrescente nas categorias e médias.

Ex.: “Grupo A está caindo 10% ao mês, se continuar, desaparecerá em X meses.”

```python
from sklearn.linear_model import LinearRegression
import numpy as np

def analisar_tendencias(df, variaveis, coluna_data='ANO_MES_PORTFOLIO', ordem_meses=None):
    if ordem_meses is None:
        ordem_meses = sorted(df[coluna_data].dropna().unique())
    
    for var in variaveis:
        print(f"\n🔶📈 Análise de tendência para: **{var}**")
        
        # Agrupa por tempo e categoria da variável
        contagem = df.groupby([coluna_data, var]).size().unstack(var).reindex(ordem_meses).fillna(0)
        
        for cat in contagem.columns:
            y = contagem[cat].values
            X = np.arange(len(y)).reshape(-1, 1)

            # Regressão linear
            model = LinearRegression().fit(X, y)
            slope = model.coef_[0]
            direction = "subindo" if slope > 0 else "descendo" if slope < 0 else "estável"

            print(f"- Categoria '{cat}' está **{direction}** a uma taxa média de **{abs(slope):.2f}** por mês")
```

```python
variaveis_categ = [
    'DSC_TIPO_PRODUTO', 'DSC_EMPRESA', 'DSC_TIPO_VENDA', 'DSC_SEGURADORA',
    'DSC_GRP_CANAL_VENDA_SEG', 'QTD_SEGS_VIGENTES', 'FLG_EXCL_PRESTAMISTA',
    'DSC_RATING_AQUISICAO_STR', 'DSC_RATING_BACEN_STR', 'DSC_RATING_STR',
    'FX_LIMITE', 'DSC_ESTADO_CIVIL', 'FAIXA_IDADE_LABEL', 'DSC_DOMINIO_EMAIL',
    'DSC_GENERO_AJUSTADO', 'DSC_REGIAO'
]

# (opcional) Criar lista ordenada de meses
ordem_meses = sorted(df['ANO_MES_PORTFOLIO'].dropna().unique())

# Executar análise
analisar_tendencias(df, variaveis=variaveis_categ, ordem_meses=ordem_meses)
```

## 🔶 2️⃣ Análise de volatilidade (oscilação)
Meça o desvio padrão ou variabilidade mês a mês para saber quais grupos são mais instáveis.

Ex.: “O grupo B não caiu tanto, mas oscila muito — risco de perda imprevisível.”

```python
def analisar_volatilidade(df, variaveis, coluna_data='ANO_MES_PORTFOLIO', ordem_meses=None, top_n=5):
    if ordem_meses is None:
        ordem_meses = sorted(df[coluna_data].dropna().unique())

    for var in variaveis:
        print(f"\n🔶📊 Análise de volatilidade para: **{var}**")
        
        # Agrupa por mês e categoria
        contagem = df.groupby([coluna_data, var]).size().unstack(var).reindex(ordem_meses).fillna(0)
        
        # Calcula desvio padrão para cada categoria
        desvios = contagem.std().sort_values(ascending=False)

        print(f"Top {top_n} categorias mais voláteis em '{var}':")
        for cat, std in desvios.head(top_n).items():
            print(f"- Categoria '{cat}': desvio padrão de **{std:.2f} adesões** por mês")
```

```python
analisar_volatilidade(
    df,
    variaveis=variaveis_categ,       # lista de variáveis já definida
    ordem_meses=ordem_meses,         # mesma lista usada na análise de tendência
    top_n=5                          # mostra as 5 categorias mais instáveis
)
```
