```python
import pandas as pd
import numpy as np
import random

# Configura√ß√µes b√°sicas
np.random.seed(42)
random.seed(42)

# Par√¢metros
safras = [
    '202401','202402','202403','202404','202405','202406',
    '202407','202408','202409','202410','202411','202412',
    '202501','202502','202503','202504'
]
dados_safra = [
    (1000,34,39),(995,50,36),(1009,18,49),(978,22,30),
    (970,21,44),(947,10,25),(932,41,40),(933,24,58),
    (899,40,38),(901,32,20),(913,30,27),(916,39,57),
    (898,47,48),(897,20,30),(887,33,10),(910,40,57)
]

# Vari√°veis demogr√°ficas simuladas
generos = ['Masculino', 'Feminino', 'Outro']
ra√ßas = ['Branco', 'Negro', 'Pardo', 'Amarelo', 'Ind√≠gena']
senioridades = ['J√∫nior', 'Pleno', 'S√™nior', 'Lideran√ßa']
√°reas = ['TI', 'RH', 'Financeiro', 'Opera√ß√µes', 'Marketing', 'Vendas']
localiza√ß√µes = ['S√£o Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre', 'Recife']

# Base geral
base_final = pd.DataFrame()
id_counter = 1
ativos = pd.DataFrame()

for idx, safra in enumerate(safras):
    qtd_inicial, novos, sa√≠das = dados_safra[idx]
    
    # Carregar popula√ß√£o anterior
    if idx == 0:
        ativos_safra = pd.DataFrame({
            'ID': range(id_counter, id_counter + qtd_inicial),
            'Data_Entrada': pd.to_datetime(f'{safra}01', format='%Y%m%d'),
            'Status': 'Ativo'
        })
        id_counter += qtd_inicial
    else:
        ativos_safra = ativos.copy()
    
    # Simular novos entrantes
    novos_ids = range(id_counter, id_counter + novos)
    novos_func = pd.DataFrame({
        'ID': novos_ids,
        'Data_Entrada': pd.to_datetime(f'{safra}01', format='%Y%m%d'),
        'Status': 'Ativo'
    })
    id_counter += novos
    ativos_safra = pd.concat([ativos_safra, novos_func], ignore_index=True)
    
    # Simular sa√≠das
    sa√≠das_ids = np.random.choice(ativos_safra['ID'], sa√≠das, replace=False)
    ativos_safra.loc[ativos_safra['ID'].isin(sa√≠das_ids), 'Status'] = 'Inativo'
    ativos_safra['Data_Saida'] = np.where(
        ativos_safra['ID'].isin(sa√≠das_ids),
        pd.to_datetime(f'{safra}28', format='%Y%m%d'),
        pd.NaT
    )
    
    # Adicionar demografia
    if 'Genero' not in ativos_safra.columns:
        ativos_safra['Genero'] = np.random.choice(generos, len(ativos_safra))
        ativos_safra['Idade'] = np.random.randint(18, 60, len(ativos_safra))
        ativos_safra['Ra√ßa'] = np.random.choice(ra√ßas, len(ativos_safra))
        ativos_safra['Senioridade'] = np.random.choice(senioridades, len(ativos_safra))
        ativos_safra['√Årea'] = np.random.choice(√°reas, len(ativos_safra))
        ativos_safra['Localiza√ß√£o'] = np.random.choice(localiza√ß√µes, len(ativos_safra))
    
    # Marcar safra
    ativos_safra['Safra'] = safra
    base_final = pd.concat([base_final, ativos_safra], ignore_index=True)
    
    # Atualizar ativos para pr√≥xima safra
    ativos = ativos_safra[ativos_safra['Status'] == 'Ativo'].copy()

# Resetar index final
base_final.reset_index(drop=True, inplace=True)
```

### üéØ Objetivo

Entender a composi√ß√£o demogr√°fica da for√ßa de trabalho ativa m√™s a m√™s (g√™nero, ra√ßa, senioridade, etc.)

Comparar o perfil de quem entra, sai e permanece

Apoiar o RH na manuten√ß√£o da diversidade, observando poss√≠veis distor√ß√µes ou tend√™ncias preocupantes

1Ô∏è‚É£ **Preparar a base anal√≠tica**

Crie 3 subconjuntos mensais, para cada safra:

* entrantes: pessoas cuja Data_Entrada seja dentro da safra

* sa√≠das: pessoas cuja Data_Saida seja dentro da safra

* ativos: todos com Status == 'Ativo' naquela safra

```python
def filtrar_perfis_por_safra(df, safra):
    entrantes = df[(df['Safra'] == safra) & (df['Data_Entrada'].dt.strftime('%Y%m') == safra)]
    saidas = df[(df['Safra'] == safra) & (df['Data_Saida'].notnull()) & (df['Data_Saida'].dt.strftime('%Y%m') == safra)]
    ativos = df[(df['Safra'] == safra) & (df['Status'] == 'Ativo')]
    return entrantes, saidas, ativos
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ordenar safras
safras_ordenadas = sorted(df['Safra'].unique())
ultimo_mes = safras_ordenadas[-1]
mes_anterior = safras_ordenadas[-2]

# Pegar dados
_, _, ativos_anterior = filtrar_perfis_por_safra(df, mes_anterior)
entrantes, saidas, ativos_atual = filtrar_perfis_por_safra(df, ultimo_mes)

# Valores
ativos_ini = len(ativos_anterior)
entradas = len(entrantes)
saidas = len(saidas)
ativos_fim = len(ativos_atual)

# Construir s√©rie para cascata
categorias = ['Ativos M√™s Anterior', 'Entradas', 'Sa√≠das', 'Ativos M√™s Atual']
valores = [ativos_ini, entradas, -saidas, ativos_fim]

# Calcular bases: acumula nas intermedi√°rias, final vai para zero
bases = [0, ativos_ini, ativos_ini + entradas, 0]

# === DEFINIR CORES DAS BARRAS ===
# Azul para totais, verde para entradas (positivo), vermelho para sa√≠das (negativo)
cores_barras = ['#1f77b4',  # Ativos anterior (azul)
                '#2ca02c',  # Entradas (verde)
                '#d62728',  # Sa√≠das (vermelho)
                '#1f77b4']  # Ativos atual (azul)
# Se quiser mudar, altere os c√≥digos hexadecimais acima

# Plot
fig, ax = plt.subplots(figsize=(10,6))

for i, (cat, val, base, color) in enumerate(zip(categorias, valores, bases, cores_barras)):
    ax.bar(cat, val, bottom=base, color=color)

# Linha acumulada para visualiza√ß√£o
acumulado = [ativos_ini, ativos_ini + entradas, ativos_ini + entradas - saidas, ativos_fim]
ax.plot(categorias, acumulado, color='black', marker='o', linestyle='--', label='Evolu√ß√£o Acumulada')

# R√≥tulos com espa√ßo extra (ajuste multiplicador conforme necessidade)
for i, (cat, total) in enumerate(zip(categorias, acumulado)):
    ax.text(i, total + max(acumulado)*0.05, f'{int(total)}', ha='center', va='bottom', fontweight='bold')

# Ajustes visuais
ax.set_title(f'Cascata Ativos | {mes_anterior} ‚Üí {ultimo_mes}')
ax.set_ylabel('N√∫mero de Funcion√°rios')
ax.set_ylim(0, max(acumulado)*1.3)  # Aumenta espa√ßo superior para os r√≥tulos
plt.legend()
plt.tight_layout()
plt.savefig('output_imgs/cascata_ativos.png')
plt.show()
```

```python
# √öltimo m√™s
ultimo_mes = sorted(df['Safra'].unique())[-1]

# Fun√ß√£o para gerar resumo e frase
def gerar_frase_resumo(subset_df, label):
    resumo = {}
    for var in variaveis:
        dist = subset_df[var].value_counts(normalize=True).round(3) * 100
        if not dist.empty:
            resumo[var] = dist.to_dict()
    
    frase = f"No m√™s {ultimo_mes}, os {label} s√£o majoritariamente "
    dominantes = []
    for var, dist in resumo.items():
        top_categoria = max(dist, key=dist.get)
        top_valor = dist[top_categoria]
        dominantes.append(f"{var} **{top_categoria}** ({top_valor:.1f}%)")
    
    frase += "; ".join(dominantes) + "."
    return frase, resumo

# Subsets
ativos_ult_mes = df[(df['Safra'] == ultimo_mes) & (df['Status'] == 'Ativo')]
entrantes_ult_mes = df[(df['Safra'] == ultimo_mes) & (df['Data_Entrada'].dt.strftime('%Y%m') == ultimo_mes)]
saidas_ult_mes = df[(df['Safra'] == ultimo_mes) & (df['Data_Saida'].notnull()) & (df['Data_Saida'].dt.strftime('%Y%m') == ultimo_mes)]

# Vari√°veis categ√≥ricas
variaveis = [col for col in df.columns if df[col].dtype == 'object' and col not in ['ID', 'Status', 'Safra']]

# Gerar frases e armazenar resumos
frase_ativos, resumo_ativos = gerar_frase_resumo(ativos_ult_mes, 'clientes ativos')
frase_entrantes, resumo_entrantes = gerar_frase_resumo(entrantes_ult_mes, 'entrantes')
frase_saidas, resumo_saidas = gerar_frase_resumo(saidas_ult_mes, 'sa√≠das')

# Imprimir frases
print("‚úÖ Frase resumo - Ativos:")
print(frase_ativos)

print("\n‚úÖ Frase resumo - Entrantes:")
print(frase_entrantes)

print("\n‚úÖ Frase resumo - Sa√≠das:")
print(frase_saidas)
```

üåü A) Descobrir perfis ocultos
Quais grupos naturais existem na for√ßa de trabalho?

Exemplo: jovens especialistas em tecnologia, mulheres s√™niores em RH, homens maduros em opera√ß√µes.

A√ß√£o: Criar a√ß√µes espec√≠ficas para cada grupo (desenvolvimento, reten√ß√£o, engajamento).

üåü B) Identificar grupos de risco ou prioridade
Depois de formar os clusters, analise os indicadores cr√≠ticos (turnover, performance, engajamento, absente√≠smo) por grupo.

Exemplo: Grupo 2 tem 30% mais sa√≠das que os outros clusters.

A√ß√£o: Priorizar diagn√≥sticos, planos de reten√ß√£o, refor√ßo de carreira para esses clusters.

üåü C) Direcionar iniciativas personalizadas
Em vez de ‚Äúuma pol√≠tica para todos‚Äù, voc√™ entende quem precisa de qu√™.

Exemplo:

Cluster A ‚Üí jovens talentos ‚Üí mais treinamento t√©cnico.

Cluster B ‚Üí s√™niores ‚Üí mais desafios, menos rotinas.

Cluster C ‚Üí lideran√ßa ‚Üí refor√ßo em gest√£o de pessoas.

üåü D) Alimentar outros modelos
Os clusters podem virar uma vari√°vel categ√≥rica nos modelos preditivos (ex.: risco de sa√≠da, probabilidade de promo√ß√£o).

A√ß√£o: Melhorar a acur√°cia dos modelos usando segmenta√ß√µes reais da popula√ß√£o.

‚úÖ Checklist para aplicar clustering corretamente
‚úî Use uma linha por indiv√≠duo, n√£o por m√™s.
‚úî Defina bem quais vari√°veis entram (evite misturar dados fixos com dados que mudam m√™s a m√™s sem cuidado).
‚úî Experimente diferentes n√∫meros de clusters (n_clusters).
‚úî Depois de formar os clusters, analise qualitativamente (n√£o s√≥ os n√∫meros) ‚Üí quem est√° em cada grupo? Quais caracter√≠sticas s√£o comuns?
‚úî Valide com o time de RH: os grupos fazem sentido no neg√≥cio?







```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Base principal
df = base_final.copy()

# √öltima linha por ID (mais atual)
df_sorted = df.sort_values(['ID', 'Safra'])
df_last = df_sorted.groupby('ID').last().reset_index()

# Calcular meses ativos
meses_ativos = df[df['Status'] == 'Ativo'].groupby('ID').size().reset_index(name='Meses_Ativo')

# Calcular tempo total de casa (dias)
df_last['Data_Saida'] = pd.to_datetime(df_last['Data_Saida'])
df_last['Data_Entrada'] = pd.to_datetime(df_last['Data_Entrada'])
df_last['Tempo_Casa'] = (df_last['Data_Saida'].fillna(pd.Timestamp.today()) - df_last['Data_Entrada']).dt.days

# Juntar meses ativos
df_final = df_last.merge(meses_ativos, on='ID', how='left')

# === Tratamento de NaNs ===
df_final['Meses_Ativo'] = df_final['Meses_Ativo'].fillna(0)
df_final['Tempo_Casa'] = df_final['Tempo_Casa'].fillna(0)
df_final['Idade'] = df_final['Idade'].fillna(df_final['Idade'].median())

cat_cols = ['Genero', 'Ra√ßa', 'Senioridade', '√Årea', 'Localiza√ß√£o']
for col in cat_cols:
    df_final[col] = df_final[col].fillna('Desconhecido')

# === SEPARAR VARI√ÅVEIS PARA CLUSTER ===
# N√ÉO incluir Tempo_Casa nem Meses_Ativo no cluster
features_for_cluster = df_final[['Idade', 'Genero', 'Ra√ßa', 'Senioridade', '√Årea', 'Localiza√ß√£o']]

# Codificar categ√≥ricas
features_encoded = pd.get_dummies(features_for_cluster, drop_first=True)

# Padronizar
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_encoded)

# Rodar KMeans
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_final['Cluster'] = kmeans.fit_predict(features_scaled)

# === RESULTADOS ===

# Distribui√ß√£o dos clusters
print("‚úÖ Distribui√ß√£o dos clusters:")
print(df_final['Cluster'].value_counts())

# Perfil categ√≥rico por cluster
for col in cat_cols:
    print(f"\nüìä Distribui√ß√£o de {col} por cluster:")
    print(df_final.groupby('Cluster')[col].value_counts(normalize=True).round(2) * 100)

# Estat√≠sticas de Tempo_Casa e Meses_Ativo por cluster
print("\n‚è≥ Estat√≠sticas de Tempo de Casa (dias) por cluster:")
print(df_final.groupby('Cluster')['Tempo_Casa'].describe())

print("\nüìÖ Estat√≠sticas de Meses Ativos por cluster:")
print(df_final.groupby('Cluster')['Meses_Ativo'].describe())
```

```python
# 1Ô∏è‚É£ Preparar resumo descritivo autom√°tico por cluster
markdown_resumo = ""

for cluster in sorted(df_final['Cluster'].unique()):
    cluster_data = df_final[df_final['Cluster'] == cluster]
    total = len(cluster_data)
    markdown_resumo += f"\n### üîπ Cluster {cluster} (n={total})\n"
    
    for col in cat_cols:
        dominante = cluster_data[col].value_counts(normalize=True).idxmax()
        pct = cluster_data[col].value_counts(normalize=True).max() * 100
        markdown_resumo += f"- {col}: **{dominante}** ({pct:.1f}%)\n"
    
    tempo_medio = cluster_data['Tempo_Casa'].mean()
    meses_medio = cluster_data['Meses_Ativo'].mean()
    markdown_resumo += f"- Tempo m√©dio de casa: **{tempo_medio:.1f} dias**\n"
    markdown_resumo += f"- Meses ativos m√©dio: **{meses_medio:.1f} meses**\n"

print("‚úÖ Resumo para markdown:")
print(markdown_resumo)
```

```python
# 2Ô∏è‚É£ Calcular top diferen√ßas percentuais vs. total geral
diferencas_markdown = ""

for col in cat_cols:
    total_pct = df_final[col].value_counts(normalize=True)
    diferencas_markdown += f"\n### üîç Vari√°vel: {col}\n"
    
    for cluster in sorted(df_final['Cluster'].unique()):
        cluster_data = df_final[df_final['Cluster'] == cluster]
        cluster_pct = cluster_data[col].value_counts(normalize=True)
        
        diffs = (cluster_pct - total_pct).dropna().abs().sort_values(ascending=False)
        top_caract = diffs.index[0]
        delta = diffs.iloc[0] * 100
        diferencas_markdown += f"- Cluster {cluster}: **{top_caract}** se destaca (+/- {delta:.1f} p.p. vs total)\n"

print("‚úÖ Top caracter√≠sticas que diferenciam clusters:")
print(diferencas_markdown)
```

```python
import pandas as pd
import numpy as np
import random

# Configura√ß√µes b√°sicas
np.random.seed(42)
random.seed(42)

# Par√¢metros
safras = [
    '202401','202402','202403','202404','202405','202406',
    '202407','202408','202409','202410','202411','202412',
    '202501','202502','202503','202504'
]
dados_safra = [
    (1000,34,39),(995,50,36),(1009,18,49),(978,22,30),
    (970,21,44),(947,10,25),(932,41,40),(933,24,58),
    (899,40,38),(901,32,20),(913,30,27),(916,39,57),
    (898,47,48),(897,20,30),(887,33,10),(910,40,57)
]

# Vari√°veis simuladas
generos = ['Masculino', 'Feminino', 'Outro']
ra√ßas = ['Branco', 'Negro', 'Pardo', 'Amarelo', 'Ind√≠gena']
senioridades = ['J√∫nior', 'Pleno', 'S√™nior', 'Lideran√ßa']
√°reas = ['TI', 'RH', 'Financeiro', 'Opera√ß√µes', 'Marketing', 'Vendas']
localiza√ß√µes = ['S√£o Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre', 'Recife']
tipos_contrato = ['CLT', 'PJ']
tipos_saida = ['Volunt√°ria', 'Involunt√°ria']

# Base final
df = pd.DataFrame()
id_counter = 1
ativos = pd.DataFrame()

for idx, safra in enumerate(safras):
    qtd_inicial, novos, sa√≠das = dados_safra[idx]
    
    if idx == 0:
        ativos_safra = pd.DataFrame({
            'ID': range(id_counter, id_counter + qtd_inicial),
            'Data_Entrada': pd.to_datetime(f'{safra}01', format='%Y%m%d'),
            'Status': 'Ativo'
        })
        id_counter += qtd_inicial
    else:
        ativos_safra = ativos.copy()
    
    novos_ids = range(id_counter, id_counter + novos)
    novos_func = pd.DataFrame({
        'ID': novos_ids,
        'Data_Entrada': pd.to_datetime(f'{safra}01', format='%Y%m%d'),
        'Status': 'Ativo'
    })
    id_counter += novos
    ativos_safra = pd.concat([ativos_safra, novos_func], ignore_index=True)
    
    # Simular sa√≠das
    sa√≠das_ids = np.random.choice(ativos_safra['ID'], sa√≠das, replace=False)
    ativos_safra.loc[ativos_safra['ID'].isin(sa√≠das_ids), 'Status'] = 'Inativo'
    ativos_safra['Data_Saida'] = np.where(
        ativos_safra['ID'].isin(sa√≠das_ids),
        pd.to_datetime(f'{safra}28', format='%Y%m%d'),
        pd.NaT
    )
    ativos_safra['Tipo_Saida'] = 'Nenhuma'  # default para todos
    if sa√≠das > 0:
        saidas_tipos = np.random.choice(tipos_saida, sa√≠das)
        ativos_safra.loc[ativos_safra['ID'].isin(sa√≠das_ids), 'Tipo_Saida'] = saidas_tipos
    
    # Adicionar vari√°veis demogr√°ficas e organizacionais
    if 'Genero' not in ativos_safra.columns:
        n = len(ativos_safra)
        ativos_safra['Genero'] = np.random.choice(generos, n)
        ativos_safra['Idade'] = np.random.randint(18, 60, n)
        ativos_safra['Ra√ßa'] = np.random.choice(ra√ßas, n)
        ativos_safra['Senioridade'] = np.random.choice(senioridades, n)
        ativos_safra['√Årea'] = np.random.choice(√°reas, n)
        ativos_safra['Localiza√ß√£o'] = np.random.choice(localiza√ß√µes, n)
        ativos_safra['PCD'] = np.random.choice([True, False], n, p=[0.1, 0.9])
        ativos_safra['Tipo_Contrato'] = np.random.choice(tipos_contrato, n, p=[0.85, 0.15])
        ativos_safra['FTE'] = np.random.uniform(0.5, 1.0, n)
        ativos_safra['Custo_Total'] = np.random.normal(8000, 2000, n)
        ativos_safra['Performance'] = np.random.choice(['Alta', 'M√©dia', 'Baixa'], n, p=[0.2, 0.6, 0.2])
        ativos_safra['Metas_Atingidas'] = np.random.uniform(60, 110, n)
        ativos_safra['Num_Reportes'] = np.random.randint(0, 10, n)
        ativos_safra['Talento_Critico'] = np.random.choice([True, False], n, p=[0.15, 0.85])
        ativos_safra['Churn_Risk_Score'] = np.random.uniform(0, 1, n)
        ativos_safra['eNPS'] = np.random.randint(-100, 100, n)
        ativos_safra['Engajamento'] = np.random.uniform(0, 100, n)
        ativos_safra['Survey_Respondido'] = np.random.choice([True, False], n, p=[0.7, 0.3])
        ativos_safra['Sentimento'] = np.random.choice(['Positivo', 'Neutro', 'Negativo'], n, p=[0.6, 0.3, 0.1])
        ativos_safra['Alinhamento_Cultural'] = np.random.uniform(0, 100, n)
        ativos_safra['Horas_Treinamento'] = np.random.uniform(0, 40, n)
        ativos_safra['Taxa_Conclusao_Treinamento'] = np.random.uniform(0, 1, n)
        ativos_safra['Plano_Desenvolvimento_Ativo'] = np.random.choice([True, False], n, p=[0.4, 0.6])
        ativos_safra['Promovido'] = np.random.choice([True, False], n, p=[0.1, 0.9])
        ativos_safra['Mov_Lateral'] = np.random.choice([True, False], n, p=[0.05, 0.95])
    
    # For√ßar preenchimento sem nulos (caso venha quebrado)
    ativos_safra.fillna({
        'Tipo_Saida': 'Nenhuma',
        'PCD': False,
        'Tipo_Contrato': 'CLT',
        'FTE': 1.0,
        'Custo_Total': 8000,
        'Performance': 'M√©dia',
        'Metas_Atingidas': 80,
        'Num_Reportes': 0,
        'Talento_Critico': False,
        'Churn_Risk_Score': 0.2,
        'eNPS': 0,
        'Engajamento': 50,
        'Survey_Respondido': False,
        'Sentimento': 'Neutro',
        'Alinhamento_Cultural': 50,
        'Horas_Treinamento': 0,
        'Taxa_Conclusao_Treinamento': 0,
        'Plano_Desenvolvimento_Ativo': False,
        'Promovido': False,
        'Mov_Lateral': False
    }, inplace=True)
    
    # Marcar safra
    ativos_safra['Safra'] = safra
    df = pd.concat([df, ativos_safra], ignore_index=True)
    
    # Atualizar ativos para pr√≥xima safra
    ativos = ativos_safra[ativos_safra['Status'] == 'Ativo'].copy()

# Resetar index final
df.reset_index(drop=True, inplace=True)
```
